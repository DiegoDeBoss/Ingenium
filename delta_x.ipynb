```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-10,10,201)
```

## Definizione della funzione $f(x)$

Qui definisco la funzione da utilizzare

```python
def f(x):
    return(x**3)
```

# Funzione pendenza $m(x)$

```python
def m(x,dx):
    return (f(x+dx)-f(x))/(dx)
```

```python
plt.plot(x, f(x), label='f(x)')
for dx in [2,1.5,1,0.5,1e-9]:
    plt.plot(x, m(x,dx), label='m(x, dx={dx})')
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x) e m(x) per x da -10 a 10')
plt.grid(True)
plt.ylim(-10, 10)
plt.legend()
plt.show()
```
